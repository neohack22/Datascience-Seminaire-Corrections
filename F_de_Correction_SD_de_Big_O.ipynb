# Notation de Grand O

Dans cette leçon, nous allons examiner le fonctionnement de la syntaxe de la notation de Grand O et comment nous pouvons décrire des algorithmes en utilisant cette notation !

Nous avons déjà discuté des fonctions ci-dessous :

```python
def sum1(n):
    '''
    Take an input of n and return the sum of the numbers from 0 to n
    '''
    final_sum = 0
    for x in xrange(n+1):
        final_sum += x

    return final_sum
```

```python
def sum2(n):
    """
    Take an input of n and return the sum of the numbers from 0 to n
    """
    return (n*(n+1))/2
```

Maintenant, nous voulons développer une notation pour comparer objectivement l'efficacité de ces deux algorithmes. Un bon point de départ serait de comparer le nombre d'affectations que chaque algorithme effectue.

La fonction **sum1** originale créera une affectation **n+1** fois, nous pouvons le voir à partir de la fonction basée sur la plage. Cela signifie qu'elle affectera la variable final_sum n+1 fois. Nous pouvons alors dire que pour un problème de taille n (dans ce cas, juste un nombre n), cette fonction prendra 1+n étapes.

Cette notation **n** nous permet de comparer des solutions et des algorithmes par rapport à la taille du problème, car sum1(10) et sum1(100000) prendraient des temps très différents pour s'exécuter, mais utiliseraient le même algorithme. Nous pouvons également noter que lorsque n devient très grand, le **+1** n'aura pas beaucoup d'effet. Commençons donc à réfléchir à la manière de construire une syntaxe pour cette notation.
________

Maintenant, nous allons réfléchir à la façon dont nous pouvons formaliser cette notation et cette idée.

La notation Big-O décrit *la vitesse à laquelle le temps d'exécution augmentera par rapport à l'entrée lorsque l'entrée devient arbitrairement grande*.

Examinons de plus près certains de ces points :

* N'oubliez pas que nous voulons comparer la vitesse à laquelle le temps d'exécution augmentera, et non comparer les temps d'exécution exacts, car ceux-ci peuvent varier en fonction du matériel.

* Puisque nous voulons comparer pour une variété de tailles d'entrée, nous ne nous préoccupons que de la croissance du temps d'exécution *par rapport* à l'entrée. C'est pourquoi nous utilisons **n** pour la notation.

* Lorsque n devient arbitrairement grand, nous ne nous soucions que des termes qui croîtront le plus rapidement à mesure que n devient grand, à ce stade, l'analyse Big-O est également connue sous le nom d'**analyse asymptotique**.

Maintenant, traçons le temps d'exécution en fonction du Big-O pour comparer les temps d'exécution. Nous utiliserons une simple [matplotlib](http://matplotlib.org/) pour le tracé ci-dessous. (Ne vous inquiétez pas de la façon d'utiliser matplotlib, cela n'est pas pertinent pour cette partie).

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

# Set up runtime comparisons
n = np.linspace(1,10,1000)
labels = ['Constant', 'Logarithmique', 'Linéaire', 'Log Linéaire', 'Quadratique', 'Cubique', 'Exponentiel']
big_o = [np.ones(n.shape),np.log(n),n,n*np.log(n),n**2,n**3,2**n]

# Plot setup
plt.figure(figsize=(12,10))
plt.ylim(0,50)

for i in range(len(big_o)):
    plt.plot(n,big_o[i],label = labels[i])


plt.legend(loc=0)
plt.ylabel('Temps d execution relatif')
plt.xlabel('n')
```

Notez à quel point l'efficacité d'un Big-O peut faire une différence pour une même valeur de n par rapport au temps d'exécution prévu ! Il est clair que nous voulons choisir des algorithmes qui évitent tout comportement exponentiel, quadratique ou cubique !
